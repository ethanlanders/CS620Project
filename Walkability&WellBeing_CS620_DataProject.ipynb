##  Walkability Score Dataset

This is our primary dataset, which was obtained from [here](https://a816-dohbesp.nyc.gov/IndicatorPublic/beta/data-explorer/walking-driving-and-cycling/?id=2133#display=summary).  

The dataset scores the walkability of most New York City neighbhorhoods from -3 to 9 as of 2011.  We will compare each neighborhood listed in this dataset for walkability with the same neighborhoods listed in other happiness index datasets to find correlations and trends.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data into .csv
df1 = pd.read_csv('https://raw.githubusercontent.com/ethanlanders/Walkability-And-Well-Being-Analysis-CS620/main/Datasets/Walkability%20Index/NYC%20EH%20Data%20Portal%20-%20Walkability%20index%20(filtered).csv')

# Rename columns for clarity
df1.rename(columns={'Score': 'Walkability Score', 'Geography': "NYC Neighborhood"}, inplace=True)

# Modify score column, which currently ranges from -3 to 9, to 0 to 13
df1['Walkability Score'] += 4

# Sort dataframe from highest walkability to lowest walkability for visualization.
sorted_df1 = df1.sort_values(by='Walkability Score', ascending=False)

# Create diagram
plt.figure(figsize=(14, 8))

# Histogram
plt.subplot(1, 2, 1)
plt.barh(sorted_df1['NYC Neighborhood'], sorted_df1['Walkability Score'], alpha=0.7)
plt.xlabel('Walkability Score')
plt.ylabel('NYC Neighborhoods')
plt.title('The Walkability of NYC Neighborhoods')

# Boxplot
plt.subplot(1, 2, 2, aspect=2)
boxplot = sns.boxplot(x=sorted_df1['Walkability Score'])
plt.title("Distribution of the Walkability of NYC Neighborhoods")

# Remove y-axis label from the boxplot
boxplot.set(xlabel='Walkability Score')

# Remove ticks from the bottom axis of the boxplot
boxplot.tick_params(axis='y', bottom=False)

# Output the plots
plt.tight_layout()
plt.show()

# Save the modified data frame to a new .csv file
df1.to_csv('df1_modified.csv', index=False)
```

###Correlation Between Walkability and Percentage of People Living in Concentrated Poverty in NYC Neighborhoods

```python
# Assuming you have columns 'Walkability Score' and 'Poverty Rate %' in your DataFrame
weight_walkability = 0.5
weight_poverty = 0.5

# Normalize factors to [0, 1]
min_walkability = comparison_df['Walkability Score'].min()
max_walkability = comparison_df['Walkability Score'].max()
comparison_df['Normalized Walkability'] = (comparison_df['Walkability Score'] - min_walkability) / (max_walkability - min_walkability)

min_poverty = comparison_df['Poverty Rate %'].min()
max_poverty = comparison_df['Poverty Rate %'].max()
comparison_df['Normalized Poverty'] = (comparison_df['Poverty Rate %'] - min_poverty) / (max_poverty - min_poverty)

# Calculate the combined score
comparison_df['Combined Score'] = (
    weight_walkability * comparison_df['Normalized Walkability'] +
    weight_poverty * comparison_df['Normalized Poverty']
)

# Normalize the combined score to [0, 1]
min_combined = comparison_df['Combined Score'].min()
max_combined = comparison_df['Combined Score'].max()
comparison_df['Normalized Combined Score'] = (comparison_df['Combined Score'] - min_combined) / (max_combined - min_combined)
```

Above, a normalized combined score was calculated for each NYC neighborhood based on the poverty rate and walkability score.  This normalized combined score is then used to sort the following visualization's legend and assign each NYC neighborhood a color on the legend from red to blue.

```python
# Selecting relevant columns from the DataFrame for visualization
poverty_walkability = comparison_df[["NYC Neighborhood", "Walkability Score", "Poverty Rate %", "Normalized Combined Score"]]

# Sort the data frame by Combined Score in descending order
poverty_walkability = poverty_walkability.sort_values(by='Normalized Combined Score', ascending=False)

# Generating a color palette that goes from dark green to dark red
num_neighborhoods = poverty_walkability['NYC Neighborhood'].nunique()
palette = sns.color_palette("RdBu_r", num_neighborhoods)

# Creating a scatter plot with the filtered dataset
plt.figure(figsize=(12, 8))
scatter = sns.scatterplot(data=poverty_walkability,
                          x='Walkability Score',     # x-axis represents Walkability Score
                          y='Poverty Rate %',        # y-axis represents Poverty Rate %
                          hue='NYC Neighborhood',    # Colors represent different NYC Neighborhoods
                          palette=palette,           # Color palette for better visibility
                          s=200)                     # Marker size

# Adding a legend with specified title and position
plt.legend(title='NYC Neighborhood', loc='upper right', bbox_to_anchor=(1.35, 1))

# Adding labels and title to the plot
plt.title('Relationship Between Walkability and Poverty Rates in NYC Neighborhoods')
plt.xlabel('Walkability Score')
plt.ylabel('Concentrated Poverty Rate Percentage (%)')

# Displaying grid on the plot for better readability
plt.grid(True)

# Showing the plot
plt.show()
```

```python
# Assuming you have columns 'Walkability Score' and 'Obesity Rate %' in your DataFrame
weight_walkability = 0.5
weight_obesity = 0.5

# Normalize factors to [0, 1]
min_walkability = comparison_df['Walkability Score'].min()
max_walkability = comparison_df['Walkability Score'].max()
comparison_df['Normalized Walkability'] = (comparison_df['Walkability Score'] - min_walkability) / (max_walkability - min_walkability)

min_obesity = comparison_df['Obesity Rate %'].min()
max_obesity = comparison_df['Obesity Rate %'].max()
comparison_df['Normalized Obesity'] = (comparison_df['Obesity Rate %'] - min_obesity) / (max_obesity - min_obesity)

# Calculate the combined score
comparison_df['Combined Score'] = (
    weight_walkability * comparison_df['Normalized Walkability'] +
    weight_obesity * comparison_df['Normalized Obesity']
)

# Normalize the combined score to [0, 1]
min_combined = comparison_df['Combined Score'].min()
max_combined = comparison_df['Combined Score'].max()
comparison_df['Normalized Combined Score'] = (comparison_df['Combined Score'] - min_combined) / (max_combined - min_combined)
```

Above, a normalized combined score was calculated for each NYC neighborhood based on the obesity rate and walkability score.  This normalized combined score is then used to sort the following visualization's legend and assign each NYC neighborhood a color on the legend from red to blue.

```python
# Selecting relevant columns from the DataFrame for visualization
obesity_walkability = comparison_df[["NYC Neighborhood", "Walkability Score", "Obesity Rate %", "Normalized Combined Score"]]

# Sorting the DataFrame based on 'Walkability Score'
obesity_walkability = obesity_walkability.sort_values(by='Normalized Combined Score', ascending=False)

# Generating a color palette that goes from dark green to dark red
num_neighborhoods = poverty_walkability['NYC Neighborhood'].nunique()
palette = sns.color_palette("RdBu_r", num_neighborhoods)

# Creating a scatter plot with the filtered dataset
plt.figure(figsize=(12, 8))
scatter = sns.scatterplot(data=obesity_walkability,
                          x='Walkability Score',     # x-axis represents Walkability Score
                          y='Obesity Rate %',        # y-axis represents Poverty Rate %
                          hue='NYC Neighborhood',    # Colors represent different NYC Neighborhoods
                          palette=palette,         # Color palette for better visibility
                          s=200)                     # Marker size

# Adding a legend with specified title and position
plt.legend(title='NYC Neighborhood', loc='upper right', bbox_to_anchor=(1.35, 1))

# Adding labels and title to the plot
plt.title('Relationship Between Walkability and Obesity Rates in NYC Neighborhoods')
plt.xlabel('Walkability Score')
plt.ylabel('Obesity Rate Percentage (%)')

# Displaying grid on the plot for better readability
plt.grid(True)

# Showing the plot
plt.show()
```
