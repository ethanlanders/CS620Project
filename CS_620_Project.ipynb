### ‚úÖ Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data into .csv
df = pd.read_csv('https://raw.githubusercontent.com/ethanlanders/CS620Project/main/NYC%20EH%20Data%20Portal%20-%20Walkability%20index%20(filtered).csv')

# Modify score column, which currently ranges from -3 to 9, to 0 to 13
df['Score'] += 4

# Sort dataframe from highest walkability to lowest walkability for visualization.
sorted_df = df.sort_values(by='Score', ascending=True)

# Create diagram
plt.figure(figsize=(10, 7))

# Histogram
plt.subplot(1, 2, 1)
plt.barh(sorted_df['Geography'], sorted_df['Score'], color='blue', alpha=0.7)
plt.xlabel('Walkability Score')
plt.ylabel('Neighborhood')

# Boxplot
plt.subplot(1, 2, 2)
sorted_df.boxplot(column='Score')
plt.title("New York Walkability")

# Output the plots
plt.tight_layout()
plt.show()
```

### ‚úÖVisualization

```python
# Poor_Population_Living_in_Concentrated_Poverty

import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV data into a DataFrame
df1 = pd.read_csv("https://raw.githubusercontent.com/ethanlanders/CS620Project/main/Datasets/Concentrated%20Poverty/Cleaned/Poor_Population_Living_in_Concentrated_Poverty_Cleaned.csv")

# Define a condition to filter rows where 'Data Format' is 'Number' and 'Time Frame' is 2020
condition = (df1['Data Format'] == 'Number') & (df1['Time Frame'] == 2020)

# Convert and round the 'Data' column to integers for the selected rows
df1.loc[condition, 'Data'] = df1.loc[condition, 'Data'].astype(int)

# Select specific columns ('Location', 'Time Frame', 'Data') and display the result
result = df1.loc[condition, ['Location', 'Time Frame', 'Data']]

# Sort the result by the 'Data' column in ascending order
sorted_result = result.sort_values(by='Data', ascending=False)

# Print the result without displaying the index
# print(sorted_result.to_string(index=False))

# Bar Chart
plt.figure(figsize=(6, 12))
plt.barh(sorted_result['Location'], sorted_result['Data'])
plt.ylabel('Neighborhood')
plt.xlabel('Number of Individuals')
plt.title('Poor Population Living in Concentrated Poverty in NYC')
plt.show()
```

## üößMerging the Datasetsüöß

The provided Python script performs a series of data manipulation tasks using the pandas library on three distinct datasets, represented as pandas DataFrames: `df1`, `df2`, and `df3`. `df1` is sourced from a CSV file containing a walkability index for NYC neighborhoods, `df2` from a CSV with obesity data among public elementary and middle school students, and `df3` from a CSV detailing the population living in concentrated poverty. Initially, the script creates two dictionaries, `map_df1_to_df2` and `map_df1_to_df3`, to establish consistent neighborhood naming conventions across the datasets, addressing discrepancies and ensuring that the same neighborhoods are referable in each DataFrame.

The DataFrames are then individually filtered to select specific years of interest‚Äî2011 for `df1` and `df2`, and 2020 for `df3`‚Äîand sorted by their respective neighborhood name columns. A comparison DataFrame, `comparison_csv`, is constructed to provide a side-by-side enumeration of unique neighborhood names from each DataFrame, which is then saved to a CSV file for potential further analysis. To facilitate the merging of these datasets, `df1` is enhanced with new columns that map its 'Geography' entries to the corresponding neighborhood names in `df2` and `df3` using the previously established dictionaries. Finally, the script merges `df1` with `df2` and the resulting DataFrame with `df3`, based on these new mapping columns, to create a singular DataFrame named `neighborhoods` that aggregates the neighborhood-centric data from all three original sources, thus setting the stage for comprehensive analysis.
