### ‚úÖ Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data into .csv
df1 = pd.read_csv('https://raw.githubusercontent.com/ethanlanders/Walkability-And-Well-Being-Analysis-CS620/main/Datasets/Walkability%20Index/NYC%20EH%20Data%20Portal%20-%20Walkability%20index%20(filtered).csv')

# Rename columns for clarity
df1.rename(columns={'Score': 'Walkability Score', 'Geography': "NYC Neighborhood"}, inplace=True)

# Modify score column, which currently ranges from -3 to 9, to 0 to 13
df1['Walkability Score'] += 4

# Sort dataframe from highest walkability to lowest walkability for visualization.
sorted_df1 = df1.sort_values(by='Walkability Score', ascending=True)

# Create diagram
plt.figure(figsize=(10, 7))

# Histogram
plt.subplot(1, 2, 1)
plt.barh(sorted_df1['NYC Neighborhood'], sorted_df1['Walkability Score'], color='blue', alpha=0.7)
plt.xlabel('Walkability Score')
plt.ylabel('Neighborhood')
plt.title('Distribution of Walkability Scores Across NYC Neighborhoods')

# Boxplot
plt.subplot(1, 2, 2)
sorted_df1.boxplot(column='Walkability Score')
plt.title("New York Walkability - Boxplot")

# Output the plots
plt.tight_layout()
plt.show()

# Save the modified data frame to a new .csv file
df1.to_csv('df1_modified.csv', index=False)
```

## ‚úÖMerging the Datasets

The provided Python code performs a comprehensive analysis of walkability, obesity rates, and poverty rates across different neighborhoods in New York City (NYC). Initially, it reads three separate CSV files into Pandas DataFrames, representing walkability, obesity data, and poverty data. The walkability DataFrame `df1` is sorted alphabetically based on NYC neighborhood names. The obesity DataFrame `df2` is filtered to retain only the data from the year 2011 and is then sorted alphabetically by location. Similarly, the poverty DataFrame `df3` is filtered to keep data from the year 2020 with a percentage data format and is sorted alphabetically by location.

To enhance clarity, column names in `df2` and `df3` are renamed to 'Obesity Rate %' and 'Poverty Rate %,' respectively. Decimal values in these columns are multiplied by 100 to represent percentages. The code then creates three new DataFrames `df1_neighborhoods`, `df2_neighborhoods`, `df3_neighborhoods` to ensure that each DataFrame has the same number of rows (65).

A comparison DataFrame `comparison_csv` is created to compare neighborhood names from each dataset side by side. This DataFrame is saved as a CSV file for manual inspection.

Two dictionaries `map_df1_to_df2` and `map_df1_to_df3` are then created to map neighborhood names from `df1` to `df2` and `df1` to `df3`, ensuring consistency for successful merging.

Two new columns `df1_to_df2_map` and `df1_to_df3_map` are added to `df1` to contain consistent neighborhood names between the three DataFrames. The code then merges the three DataFrames `df1`, `df2`, and `df3` into a new DataFrame `merged` using the mapped neighborhood names. Finally, a final DataFrame `comparison_df` is created by selecting relevant columns from the `merged` DataFrame, including NYC neighborhood names, walkability scores, obesity rates, and poverty rates. The result is printed for further analysis.


### ‚úÖCode
